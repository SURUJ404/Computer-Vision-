```python
import os
import cv2
from glob import glob
import matplotlib.pyplot as plt
import pandas as pd
from functools import reduce
from xml.etree import ElementTree as et
```

```python
##loading xml files
xmlfiles = glob('./data_images/*.xml')
#datacleaning
replace_text = lambda x: x.replace('\\','/')
xml_list = list(map(replace_text,xmlfiles))
```

```python
xml_list
```

```python
#step read xml files
#from each xml files we need to to extract
#filename, size(width , height) , object , (name , xmin , xmax , ymin , ymax)
def extract_text(filename):#helperfunction
    tree = et.parse(filename) #one image directory
    root = tree.getroot()
    # print(root)


    # #extract filenames
    # of the same images as above
    image_name = root.find('filename').text 
    #width =and height of the aimages
    width = root.find('size').find('width').text
    height = root.find('size').find('height').text

    objs = root.findall('object')
    parser = [] #empty list
    for obj in objs:
        obj = objs[0]
        name = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmin = bndbox.find('xmin').text
        xmax = bndbox.find('xmax').text
        ymin = bndbox.find('ymin').text
        ymax = bndbox.find('ymax').text
        parser.append([image_name , width, height, name , xmin, xmax , ymin , ymax])
   
    return parser
```

```python
parser_all = list(map(extract_text,xmlfiles))
```

```python
parser_all
```

```python
data = reduce(lambda x , y  : x+y, parser_all)
```

```python
df = pd.DataFrame(data,columns = ['filename', 'width' , 'height' , 'name' , 'xmin' , 'xmax' , 'ymin' , 'ymax'])
```

```python
##Performinf label

df['center_x'] = ((df['xmax']+df['xmin'])/2)/df['width']

df['center_x'] = ((df['ymax']+df['xmin'])/2)/df['height']

df['w'] = (df['xmax']-df['xmax'])/df['width']

df['h'] = (df['ymax']-df['ymax'])/df['height']
```

# Assign id number to object names

```python
# Label Encoding
def label_encoding(x):
    labels = {'Person':0 , 'Dog':1 , 'Car':2 , 'Bike':3 , 'Airplane':4 , 'Stingray':5 , 'Bus':6 , 'Birds':7}
    return labels[x]
```

```python
train_df['id'] = train_df['name'].apply(label_encoding)
test_df['id'] = test_df['name'].apply(label_encoding)
```
